```python
# Create a connection to the database and open a cursor
import sqlite3
conn=sqlite3.connect('my_database.sqlite')
cursor=conn.cursor()
print("Database opened successfully")
```

```python
cursor.execute('''CREATE TABLE SCHOOL
         (ID INT PRIMARY KEY     NOT NULL,
         NAME           TEXT    NOT NULL,
         AGE            INT     NOT NULL,
         ADDRESS        CHAR(50),
         MARKS          INT);''')
cursor.close()
```
